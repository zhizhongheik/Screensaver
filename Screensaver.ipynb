## The main code

This is the original code without adding anything new, The screen size here can be modified according to your device. I set 50 stars with a speed of 1 to fly outward from the center of the screen. This number can be adjusted appropriately.

````python
import pygame
import random

screen_width = screen_height = 600


pygame.init()
screen = pygame.display.set_mode((screen_width, screen_height))
done = False


number_of_stars = 50
speed =	1
stars = []

def new_star() -> list:
    
    star = [0, 0, 0, 0]

    star[0] = random.randint(-screen_width // 2, screen_width // 2)
    star[1] = random.randint(-screen_height // 2, screen_height // 2)
   
    star[2] = 256
    star[3] = 255
    
    return star


for i in range(0, number_of_stars):		
    stars.append(new_star())

while not done:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True

    screen.fill((0, 0, 0))  

    for i in range(0, number_of_stars):		
        s = stars[i]		    		
        
        # Move the star and check if it still appear
        def move_and_check(star:list) -> list:
    
            x = screen_width // 2  
            y = screen_height // 2
   
            star[0] = (star[0]*256)/star[2] + x
            star[1] = (star[1]*256)/star[2] + y

            star[2] -= speed 
    
            if star[0] < 0 or star[0] > screen_width or star[1] < 0 or star[1] > screen_height:
                star = new_star()
    
    
            if star[3] < 255:
                star[3] += 0.15
    
   
            if star[3] > 255:			
                star[3] = 255

            return star

        s = move_and_check(s)
        
        stars[i] = s
        
        # draw the star on the screen
        def draw_star(star:list) -> None:
    
            x = int(star[0])
            y = int(screen_height - star[1])  

            pygame.draw.circle(screen, (star[3], star[3], star[3]), (x, y), 3)

        draw_star(s)


    pygame.display.flip()
pygame.quit()
````

## Added new codes for meteorites and suns

I used a yellow circle placed in the upper left corner as the sun, and added meteorites like stars flying outward from the center of the screen.

````python
import pygame
import random


screen_width = screen_height = 600
number_of_stars = 50
speed = 1
stars = []
number_of_meteors = 30
meteors = []

pygame.init()
screen = pygame.display.set_mode((screen_width, screen_height))
done = False



def new_star() -> list:
    star = [0, 0, 0, 0]
    star[0] = random.randint(-screen_width // 2, screen_width // 2)
    star[1] = random.randint(-screen_height // 2, screen_height // 2)
    star[2] = 256
    star[3] = 255    
    return star


def new_meteor() -> list:
    meteor = [0, 0, 0]
    meteor[0] = random.randint(0, screen_width)
    meteor[1] = random.randint(0, screen_height)
    meteor[2] = random.uniform(1.0, 3.0)  
    return meteor


for i in range(0, number_of_stars):		
    stars.append(new_star())
for i in range(0, number_of_meteors):
    meteors.append(new_meteor())

while not done:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True

    screen.fill((0, 0, 0))

    for i in range(0, number_of_stars):		
        s = stars[i]
        
        
        def move_and_check(star:list) -> list:
            x = screen_width // 2
            y = screen_height // 2
            star[0] = (star[0]*256)/star[2] + x
            star[1] = (star[1]*256)/star[2] + y
            star[2] -= speed 
            if star[0] < 0 or star[0] > screen_width or star[1] < 0 or star[1] > screen_height:
                star = new_star()
            if star[3] < 255:
                star[3] += 0.15
            if star[3] > 255:			
                star[3] = 255
            return star

        s = move_and_check(s)
        stars[i] = s

        
        def draw_star(star:list) -> None:
            x = int(star[0])
            y = int(screen_height - star[1])
            pygame.draw.circle(screen, (star[3], star[3], star[3]), (x, y), 3)

        draw_star(s)

    for i in range(number_of_meteors):
        m = meteors[i]
        m[1] += m[2]
        if m[1] > screen_height:
            meteors[i] = new_meteor()
        else:
            pygame.draw.circle(screen, (128, 128, 128), (m[0], int(m[1])), 2)

    pygame.display.flip()

pygame.quit()
````
